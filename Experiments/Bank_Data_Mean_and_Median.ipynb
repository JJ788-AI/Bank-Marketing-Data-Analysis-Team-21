# Bank Data Means and Medians

#### Step 1: Importing pandas library to read csv data file

```python
import pandas as pd
data = pd.read_csv('../Data/processed/bank.csv')
data.head()
```

#### Step 2: Calculating the mean of multiple columns using the select_dtypes() method to filter for the numeric columns only

```python
numeric_means = data.select_dtypes(include = 'number').mean()
print(numeric_means)
```

#### Step 3: Calcultating the median of multiple columns using the select_dtypes() method to filter for the numeric columns only

```python
numeric_medians = data.select_dtypes(include = 'number').median()
print(numeric_medians)
```

#### Step 4: Import matplotlib to access the hist() method  to plot histograms of all numeric variables

```python
import matplotlib.pyplot as plt

data['age'].hist(bins = 10, edgecolor = 'black', color = 'blue' )
plt.title ('Histogram of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()

data['balance'].hist(bins = 10, edgecolor = 'black', color = 'purple')
plt.title ('Histogram of Balance')
plt.xlabel('Balance')
plt.ylabel('Frequency')
plt.show()

data['day'].hist(bins = 10, edgecolor = 'black', color = 'green')
plt.title ('Histogram of Day')
plt.xlabel('Day')
plt.ylabel('Frequency')
plt.show()

data['duration'].hist(bins = 10, edgecolor = 'black', color = 'red')
plt.title ('Histogram of Duration')
plt.xlabel('Duration')
plt.ylabel('Frequency')
plt.show()

data['campaign'].hist(bins = 10, edgecolor = 'black', color = 'grey')
plt.title ('Histogram of Campaign')
plt.xlabel('Campaign')
plt.ylabel('Frequency')
plt.show()

data['pdays'].hist(bins = 10, edgecolor = 'black', color = 'orange')
plt.title ('Histogram of Pdays')
plt.xlabel('Pdays')
plt.ylabel('Frequency')
plt.show()

data['previous'].hist(bins = 10, edgecolor = 'black', color = 'coral')
plt.title ('Histogram of Previous')
plt.xlabel('Previous')
plt.ylabel('Frequency')
plt.show()
```

#### Step 5: Import seaborn to use boxplot to get correlation between categorical target variable vs numerical predictor variables

```python
import seaborn as sns

sns.boxplot(x = 'age', y = 'outcome', data = data)
plt.title('Boxplot of Outcome by Age')
plt.show()
```


The boxplot shows that individuals who subscribed ("yes") tend to have a slightly higher median age compared to those who did not ("no"). Both groups share similar age ranges, but the "no" group includes more older-age outliers, suggesting younger and middle-aged individuals may dominate both outcomes, with a slight preference for older individuals in the "yes" category.


```python
import seaborn as sns

sns.boxplot(x = 'balance', y = 'outcome', data = data)
plt.title('Boxplot of Outcome by Balance')
plt.show()
```

The boxplot reveals that for both outcomes ("yes" and "no"), most individuals have low account balances, clustered near zero. However, clients who subscribed to the term deposit ("yes") seem to have slightly higher balances overall, with fewer outliers extending into very high balances compared to the "no" group. This indicates a possible positive association between higher account balances and the likelihood of subscription, though the majority of both groups remain concentrated in the lower balance range.

```python
import seaborn as sns

sns.boxplot(x = 'day', y = 'outcome', data = data)
plt.title('Boxplot of Outcome by Day')
plt.show()
```

The boxplot shows that the distribution of days for contacting clients (variable day) is similar between the two outcomes (yes and no). Both groups have a median around the middle of the month, with similar interquartile ranges and overall spread. This suggests that the day of the month does not have a clear or significant impact on the likelihood of a positive outcome.

```python
import seaborn as sns

sns.boxplot(x = 'campaign', y = 'outcome', data = data)
plt.title('Boxplot of Outcome by Campaign')
plt.show()
```

The boxplot illustrates the distribution of campaign counts (number of contacts) for two outcomes: "yes" and "no." The "no" outcome generally involves more contacts per campaign, as evidenced by a longer tail and more outliers at higher values. In contrast, the "yes" outcome typically requires fewer contacts, with most data concentrated in the lower campaign range and fewer outliers.

```python
import seaborn as sns

sns.boxplot(x = 'pdays', y = 'outcome', data = data)
plt.title('Boxplot of Outcome by Pdays')
plt.show()
```

The boxplot shows the distribution of pdays (number of days since a client was last contacted) for the "yes" and "no" outcomes. For the "yes" outcome, most contacts occur after a shorter duration (lower pdays), with fewer outliers at higher values. In contrast, the "no" outcome has a wider spread, with many instances across the entire range of pdays, including higher values, suggesting less correlation between short pdays and a negative outcome.

```python
import seaborn as sns

sns.boxplot(x = 'previous', y = 'outcome', data = data)
plt.title('Boxplot of Outcome by Previous')
plt.show()
```



The boxplot shows the distribution of the "previous" variable (number of prior contacts performed before this campaign and for this client) for the "yes" and "no" outcomes. For the "yes" outcome, most instances have fewer previous contacts, with a smaller interquartile range and fewer outliers extending to higher values. In contrast, the "no" outcome has a wider distribution, with previous contacts spanning a larger range and more frequent higher values. This suggests that fewer prior contacts are more likely associated with a "yes" outcome.


Note: Boxplot of duration variable was omitted due to the highimpact it has on the ouput target. As stated in the data source file - e.g., if duration=0 then y='no'. Yet, the duration is not known before a call is performed. Also, after the end of the call y is obviously known. Thus, this input should only be included for benchmark purposes and should be discarded if the intention is to have a realistic predictive model (source- https://archive.ics.uci.edu/dataset/222/bank+marketing)
